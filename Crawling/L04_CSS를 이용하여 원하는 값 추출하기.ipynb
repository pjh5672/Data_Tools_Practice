```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://news.v.daum.net/v/20190728165812603'
resp = requests.get(url)
resp.text
```

```python
soup = BeautifulSoup(resp.text)
```

```python
soup.select('h3')
```

```python
soup.select('#harmonyContainer') # id는 id명 앞에 # 을 사용해서 명시
```

```python
soup.select('div#harmonyContainer') # 앞에 div 태그를 붙일수 있음
```

```python
soup.select('#harmonyContainer p') # harmonyContainer 안의 p 태그(자손)를 모두 반환
```

```python
soup.select('#harmonyContainer > p') # harmonyContainer 안의 바로 하위 p태그(자식)만 모두 반환
```

```python
soup.select('h3.tit_view') # 클래스는 클래스 명 앞에 .을 찍어서 명시
```

```python
soup.select('.tit_view') # 클래스는 클래스 명 앞에 .을 찍어서 명시
```

```python
soup.select('h3')
```

```python
soup.select('h3[class="tit_view"]') # 대괄호 안에 class 이름을 명시해서 가져옴
```

```python
soup.select('h3[class^="t"]') # "^" 사용은 "t"로 시작하는 문자열의 클래스를 가져옴
```

```python
soup.select('h3[class$="_view"]') # "$" 사용은 "_view"로 끝나는 문자열의 클래스를 가져옴
```

```python
soup.select('h3[class*="_"]') # "*" 사용은 부분적으로 "_" 포함하는 문자열을 찾겠다
```

```python
soup.select('span.txt_info')
```

```python
soup.select('span.txt_info:nth-child(1)') # 여러개의 동일한 class를 가지는 태크콘텐츠의 경우 nth-child(n)를 통해 가져올수있음
```

```python
soup.select('span.txt_info:nth-child(2)') # 두번째 태크콘텐츠를 가져오는 경우
```
