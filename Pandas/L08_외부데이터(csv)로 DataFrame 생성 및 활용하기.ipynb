### 샘플 csv 데이터파일로 DataFrame 데이터 생성하기

```python
import numpy as np
import pandas as pd
```

- read_csv 함수 파라미터
 - sep : 각 데이터 값을 구별하기 위한 구분자(seperator) 설정
 - header : header를 무시할 경우, None  설정
 - index_col : index로 사용할 column 설정
 - usecols : 실제로 dataframe에 로딩할 columns만 설정

```python
# data 출처: https://www.kaggle.com/hesh97/titanicdataset-traincsv/data
train_data = pd.read_csv('data/train.csv')
train_data.head()
```

```python
# sep 파라미터의 기능 : 원본데이터의 구분자 기준으로 Data Frame을 생성
train_data = pd.read_csv('./data/train.csv', sep=',')
train_data
```

```python
# header 파라미터의 기능 : header 를 따로 지정 혹은 None 으로 설정가능
train_data = pd.read_csv('./data/train.csv', header=None)
train_data
```

```python
# index_col 파라미터의 기능 : 인덱스 기준을 특정 column으로 지정
train_data = pd.read_csv('./data/train.csv', index_col='PassengerId')
train_data
```

```python
# usecols 파라미터의 기능 : 필요한 Column 만 가져오기
train_data = pd.read_csv('./data/train.csv', usecols=['PassengerId', 'Survived', 'Pclass', 'Name'])
train_data
```

### DataFrame column 선택하기
 - 기본적으로 []는 column을 추출
 - 컬럼 인덱스일 경우 인덱스의 리스트 사용가능
   - 리스트를 전달할 경우 결과는 DataFrame
   - 하나의 컬럼명을 전달할 경우 결과는 Series

- 단일 컬럼 선택

```python
# Series 로 가져오기
train_data['Survived']
```

```python
# DataFrame 으로 가져오기
train_data[['Survived']]
```

- 복수 컬럼 선택

```python
train_data[['Survived', 'Age', 'Name', 'Embarked']]
```

- DataFrame Slicing

```python
train_data[:3]
```

### DataFrame row 선택하기
 - loc, iloc 함수로 row을 추출할 수 있고 ',' 을 하용하여 column도 선택가능
   - loc : 지정된 인덱스 기준으로 row 추출
   - iloc : 0 기준의 인덱스로 row 추출

```python
# loc 과 iloc 기능 차이점 확인을 위해 기존 인덱스값 변경
train_data.index = np.arange(100, 991)
train_data.head()
```

- loc 함수 사용

```python
train_data.loc[990]
```

```python
train_data.loc[[800, 900, 950, 990]]
```

- iloc 함수 사용

```python
train_data.iloc[0]
```

```python
train_data.iloc[[700, 800, 850, 890]]
```

- row, column 동시에 선택하기
  - loc, iloc 속성을 이용할 때, 콤마를 이용하여 둘 다 명시가능

```python
# loc 은 네이밍된 인덱스로 row와 column 을 동시호출 가능
train_data.loc[[986, 100, 110, 990], ['Survived', 'Name', 'Sex', 'Age']]
```

```python
# iloc 은 0 기준의 인덱스로 row와 column 을 동시호출 가능
train_data.iloc[[886, 0, 10, 890], [1, 3, 4, 5]]
```
