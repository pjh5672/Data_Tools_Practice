### 숫자데이터와 범주형 데이터의 이해
 - info 함수로 각 변수의 데이터타입 확인
 - 타입의 변경은 astype 함수를 사용
 
- 숫자형(Numerical Type) 데이터
 - 연속성을 띄는 숫자로 이루어진 데이터
- 범주형(Categorical ) 데이터
 - 연속적이지 않은 값으로 이루어진 데이터

```python
# data 출처: https://www.kaggle.com/hesh97/titanicdataset-traincsv/data
train_data = pd.read_csv('data/train.csv')
train_data.head()
```

```python
train_data.info()
# 문자열은 object로 되어있고, 숫자형은 정수의 경우 int, 실수의 경우 float 으로 표현되어있음
```

```python
# Pclass는 범주형 데이터이기 때문에 문자형으로 변환시켜준다
train_data['Pclass'] = train_data['Pclass'].astype(str)
train_data.info()
```

- 숫자형 데이터 (Age 변수)를 범주형으로 변환하기
 - 변환 로직을 함수로 만든 후, apply 함수 적용

```python
import math

def age_categorize(age):
    if math.isnan(age):
        return -1
    return math.floor(age / 10) * 10
```

```python
train_data['Age'].apply(age_categorize)
```

### 범주형 데이터 전처리하기 (one-hot encoding)
 - 범주형 데이터는 연산가능한 숫자형으로의 변환이 필요함
 - 범주형 데이터의 각 범주(category)를 column 레벨로 변경
 - 해당 범주에 속하면 1, 아니면 0 으로 채우는 인코딩 기법(one-hot encoding)
 - pandas.get_dummies 함수 사용
  - drop_first: 첫번째 카테고리 값은 사용하지 않음

```python
# 범주형 데이터에(object)로 타입이 정의된 모든 columns에 one-hot encoding 수행해서 column 개수가 증가함
pd.get_dummies(train_data)
```

```python
# 지정된 column 에 대해서만 one-hot encoding 수행
pd.get_dummies(train_data, columns=['Pclass', 'Sex', 'Embarked'])
```

```python
pd.get_dummies(train_data, columns=['Pclass', 'Sex', 'Embarked'], drop_first=True) # drop_first 기본값은 False
```

### Group by 이해하기
- GroupBy groups 속성
 - 각 그룹과 그룹에 속한 index를 dict형으로 표현

```python
# data 출처: https://www.kaggle.com/hesh97/titanicdataset-traincsv/data
df = pd.read_csv('data/train.csv')
df.head()
```

```python
class_group = df.groupby('Pclass')
class_group.groups
```

```python
gender_group = df.groupby('Sex')
gender_group.groups
```

- groupping 함수
 - 그룹데이터에 적용가능한 통계함수(NaN 제외하여 연산)
 - count, sum, mean, median, std, var, min, max 등

```python
class_group.count()
```

```python
# 등석별 생존율
class_group.mean()
```

```python
# 등석별 연령층 중위수
class_group.median()
```

```python
# 성별에 따른 생존율
gender_group.mean()
```

- 복수 columns 으로 groupping 하기
 - groupby 에 column 리스트를 전달
 - 통계함수 적용결과는 multi-index를 갖는 dataframe

```python
df.groupby(['Pclass', 'Sex']).mean()
```

```python
df.groupby(['Pclass', 'Sex']).mean().index
```

```python
df.groupby(['Pclass', 'Sex']).mean().loc[(2, 'female')]
```

- 나이대별로 생존율 구하기

```python
# Age index 를 이용해서 groupby
df.set_index('Age').groupby(level=0).mean() # multi-index 일 경우 level 이 0 -> 1 -> 2 로 증가시켜 지정해줘야함
```

```python
import math
def age_categorize(age):
    if math.isnan(age):
        return -1
    return math.floor(age/10)*10
```

```python
df.set_index('Age').groupby(age_categorize).mean()['Survived']
```

- Multi-index를 이용한 groupping

```python
df.set_index(['Pclass', 'Sex']).groupby(level=[0, 1]).mean()
```

- aggregate(집계) 함수 사용하기

```python
# 통계처리 관련 함수생성 및 로드하여 한번에 모두 적용 
df.set_index(['Pclass', 'Sex']).groupby(level=[0, 1]).aggregate([np.mean, np.median, np.max, np.min])
```

### transform 함수
 - groupby 이후 transform 함수를 사용하면 원래 index를 유지한 상태로 통계함수 적용
 - 전체데이터의 집계가 아닌 각 그룹에서의 집계를 계산
 - 따라서 새로 생성된 데이터를 원본 데이터프레임과 합치기 쉬움

```python
# data 출처: https://www.kaggle.com/hesh97/titanicdataset-traincsv/data
df = pd.read_csv('data/train.csv')
df.head()
```

```python
# 원본데이터 index 가 groupping 되어서 데이터프레임의 row 과 column 속성이 변경됨
df.groupby('Pclass').mean() 
```

```python
# 원본데이터 index 유지하고, groupping 된 index에 모두 통계함수 적용하여 반환
df.groupby('Pclass').transform(np.mean)
```

```python
# 통계처리된 데이터를 원형 데이터프레임에 합치기 쉬움
df['Age2'] = df.groupby('Pclass').transform(np.mean)['Age']
df
```

```python
df['Age3'] = df.groupby(['Pclass', 'Sex']).transform(np.mean)['Age']
df
```
