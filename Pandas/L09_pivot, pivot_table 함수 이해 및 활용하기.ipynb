### pivot, pivot_table 함수 이해 및 활용하기

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    '지역': ['서울','서울','서울','경기','경기','부산','서울','서울','부산','경기','경기', '경기'],
    '요일': ['월요일','화요일','수요일','월요일','화요일','월요일','목요일','금요일','화요일','수요일','목요일', '금요일'],
    '강수량': [100, 80, 1000, 200, 200, 100, 50, 100, 200, 100 , 50, 100],
    '강수확률': [80, 70, 90, 10, 20, 30, 50, 90, 20, 80 , 50, 10]
})
display(df)
```

- pivot 
  - dataframe의 형태 변경
  - index, column, data로 사용할 column을 명시

```python
# 지역, 요일별
df.pivot(index='지역', columns='요일')
```

```python
df.pivot(index='요일', columns='지역')
```

```python
df.pivot(index='요일', columns='지역', values='강수량')
```

- pivot_table
 - 기능적으로 pivot 과 동일하나, pivot_table은 중복데이터에 대한 처리가 가능함

```python
# 서울-월요일 중복데이터 수정 후 pivot 함수 실행하면 duplicate 오류 발생

df = pd.DataFrame({
    '지역': ['서울','서울','서울','경기','경기','부산','서울','서울','부산','경기','경기', '경기'],
    '요일': ['월요일','월요일','수요일','월요일','화요일','월요일','목요일','금요일','화요일','수요일','목요일', '금요일'],
    '강수량': [100, 80, 1000, 200, 200, 100, 50, 100, 200, 100 , 50, 100],
    '강수확률': [80, 70, 90, 10, 20, 30, 50, 90, 20, 80 , 50, 10]
})
display(df)
```

```python
df.pivot(index='요일', columns='지역')
```

```python
pd.pivot_table(df, index='요일', columns='지역', aggfunc=np.mean)
# 월요일 - 서울 지역은 강수량, 강수확률이 평균되어 반환됨
```
