### stack & unstack
- stack : 컬럼 레벨에서 인덱스 레벨로 dataframe 변경 (데이터를 쌓아올림)
- unstack : 인덱스 레벨에서 컬럼 레벨로 dataframe 변경

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    '지역': ['서울','서울','서울','경기','경기','부산','서울','서울','부산','경기','경기', '경기'],
    '요일': ['월요일','화요일','수요일','월요일','화요일','월요일','목요일','금요일','화요일','수요일','목요일', '금요일'],
    '강수량': [100, 80, 1000, 200, 200, 100, 50, 100, 200, 100 , 50, 100],
    '강수확률': [80, 70, 90, 10, 20, 30, 50, 90, 20, 80 , 50, 10]
})
display(df)
```

```python
new_df = df.set_index(['지역', '요일'])
new_df
```

```python
new_df.index
```

- unstack 호출

```python
new_df.unstack(level=0) # index level = 0 인 '지역'을 컬럼 레벨로 올림
```

```python
new_df.unstack(level=1) # index level = 1 인 '요일'을 컬럼 레벨로 올림
```

- stack 호출

```python
new_df.unstack(0).stack(0) # column level = 0 인 '강수량'과 '강수확률'이 index로 변경됨
```

```python
new_df.unstack(0).stack(1) # column level = 1 인 '지역'이 index로 변경됨
```
